```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, Imputer

df = pd.read_csv('Datas/housing.csv')
df.head()
```

```python
df["rooms_per_households"]= df["total_rooms"]/df["households"]
df["bedrooms_per_rooms"]= df["total_bedrooms"]/df["total_rooms"]
df["population_per_households"]= df["population"]/df["households"]
```

```python
encoder = LabelEncoder()
ocean_num = encoder.fit_transform(df["ocean_proximity"])
list_labels =encoder.classes_
```

```python
enc = OneHotEncoder()
t= enc.fit_transform(ocean_num.reshape(-1,1))
t.toarray()
```

```python
df_ocean = pd.DataFrame(t.toarray(), columns=list_labels)
```

```python
df_ocean.head()
```

```python
df_num = df.drop("ocean_proximity", axis=1)
```

```python
df=df_num.join(df_ocean)
```

```python
df.head()
```

```python
imputer = Imputer(strategy = "median")
imputer.fit(df)
medians = imputer.statistics_
```

```python
T = imputer.transform(df)
```

```python
df= pd.DataFrame(T, columns= df.columns)
```
