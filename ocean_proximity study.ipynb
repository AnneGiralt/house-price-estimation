```python
import pandas as pd

df = pd.read_csv('Datas/housing.csv')
```

```python
set(df.ocean_proximity)
```

```python
dico = {'<1H OCEAN' : 2 , 'INLAND': 20 , 'ISLAND': 0, 'NEAR BAY': 3, 'NEAR OCEAN' : 1}
```

```python
df.ocean_proximity.replace(dico, inplace=True)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

df.plot(kind="scatter", x="longitude", y="latitude", alpha=0.4, label="Population",
c="ocean_proximity", cmap=plt.get_cmap("jet"), colorbar=True,
)
plt.legend()
```

```python
df.plot(kind="scatter", x= "ocean_proximity", y = "median_house_value", alpha=0.005 )
```
